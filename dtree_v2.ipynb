```python

print(__doc__)
import csv
import urllib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url= "https://raw.githubusercontent.com/avirynd/Bikeshare/master/RidesMinutes.csv?token=AVmUmUm2ljKEwWgVNhcmIk3agE483Dlnks5ZF6fFwA%3D%3D"
ride_min=pd.read_csv(url)


ride_min['Result']=" "
ride_min
```

```python
ride_min['TotalMinutes'].describe()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier #This is the library that implements the Decision Tree classifier 
from sklearn.model_selection import train_test_split #used to split the dataset into training and test data
#from sklearn.tree import DecisionTreeClassifier
```

```python
print("ride_min.keys(): {}".format(ride_min.keys()))
```

```python
#print the shape of the dataset
print("Shape of data: {}".format(ride_min.TotalMinutes.shape))
```

```python
ride_min['Result_Names']=' '
#1=good
#0=bad day
ride_min.loc[ (ride_min['TotalMinutes'] < 15199), 'Result']= '0'
ride_min.loc[ (ride_min['TotalMinutes'] < 15199), 'Result_Names']= 'bad'
ride_min.loc[ (ride_min['TotalMinutes'] >= 15199), 'Result']= '1'
ride_min.loc[ (ride_min['TotalMinutes'] >= 15199), 'Result_Names']= 'good'
```

```python
ride_min
```

```python

X_train, X_test, y_train, y_test = train_test_split(
    ride_min.TotalMinutes, ride_min.Result, stratify=ride_min.Result, random_state=0)
```

```python
#build the tree using the training data
tree = DecisionTreeClassifier()
X_train=X_train.reshape(-1,1)
X_test=X_test.reshape(-1,1)
tree.fit(X_train, y_train)
#Print accuracy of the model
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
#Build the tree again setting a pre-pruning criterion for maximum depth. See how different depths affect the accuracies.
tree2 = DecisionTreeClassifier(max_depth=4)
tree2.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(tree2.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree2.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz

export_graphviz(tree, out_file="ride_min_2.dot", class_names=["bad", "good"],
                feature_names=ride_min.Result_Names, impurity=True, filled=True)


#names=np.array(['good','bad'],dtype='<U9')

#print("Sample counts per class:\n{}".format(
 #     {n: v for n, v in zip(ride_min.Result_Names, np.bincount(ride_min.Result))}))
```
