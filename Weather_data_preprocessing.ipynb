```python
print(__doc__)

import numpy as np
import scipy as sp
import pandas as pd

from sklearn.preprocessing import Imputer

# Loads the data from the two csv need for pre-processing (weather_data, trip_data)

weather_data = pd.read_csv("C:/Users/Vi/Desktop/CMPE 188/Bikeshare-master/Data/weather_data.csv")
trip_data = pd.read_csv("C:/Users/Vi/Desktop/CMPE 188/Bikeshare-master/Data/trip_data.csv")

weather_data
```

```python
# create a new column call Cloud Description. Categorized the data in cloudy cover into 5 groups. 

cloud_status = {0: "Clear", 1: "Clear", 2: "Little Cloudy", 3: "Cloudy", 4: "Cloudy", 5: "Very Cloudy", 6: "Very Cloudy", 7: "Overcast", 8: "Overcast"}

weather_data['CloudDesc'] = ""

weather_data
```

```python
# Apply the group categorization into the newly added column
weather_data['CloudDesc'] = weather_data['CloudCover '].map(cloud_status)
```

```python
# Verify that the data is correct
weather_data
```

```python
# We notice that we have a value T for some of our data entries in our precipitation column. This must be correlated to an
# integer before we can categorized the dataset. In this case T is represented by 0.001.

weather_data['PrecipitationIn '] = weather_data['PrecipitationIn '].replace('T', '0.001')

weather_data['PrecipDesc'] = ""
weather_data
```

```python
# Print and analysis the proper ranges for our dataset since it is not given.
weather_data['PrecipitationIn '].describe()
```

```python
# Convert the column of PrecipitationIn into a float because the previous entries of T made the column into a string.
weather_data['PrecipitationIn '] = weather_data['PrecipitationIn '].astype(float)

# Categorize the the values from our analysis into 5 ranges.
weather_data.loc[(weather_data['PrecipitationIn '] >= 0) & (weather_data['PrecipitationIn '] < 0.20), 'PrecipDesc'] = 'None'
weather_data.loc[(weather_data['PrecipitationIn '] >= 0.20) & (weather_data['PrecipitationIn '] < 0.50), 'PrecipDesc'] = 'Light'
weather_data.loc[(weather_data['PrecipitationIn '] >= 0.50) & (weather_data['PrecipitationIn '] < 1), 'PrecipDesc'] = 'Medium'
weather_data.loc[(weather_data['PrecipitationIn '] >= 1) & (weather_data['PrecipitationIn '] < 1.99), 'PrecipDesc'] = 'Pouring'
weather_data.loc[(weather_data['PrecipitationIn '] >= 1.99) & (weather_data['PrecipitationIn '] <= 4), 'PrecipDesc'] = 'Heavy'
```

```python
# View the results of our categorization

weather_data
```

```python
# Save the new created results into a new csv file called weather_data_categorized to be used for clustering later.

weather_data.to_csv('weather_data_categorized.csv', sep=',')
```
