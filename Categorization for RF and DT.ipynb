```python
print(__doc__)

import numpy as np
import scipy as sp
import pandas as pd

from sklearn.preprocessing import Imputer

#loads the rides minutes csv into a dataframe
rides_minutes = pd.read_csv("C:/Users/Vi/Desktop/CMPE 188/Bikeshare-master/Data/RidesMinutes.csv")
```

```python
rides_minutes
```

```python
#prints total minutes to see the min and max values
rides_minutes['TotalMinutes'].describe()
```

```python
#makes an empty column called MinCategorize
rides_minutes['MinCategorize'] = ""
```

```python
#Sets the 10 different categorizes depending on range
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 0) & (rides_minutes['TotalMinutes'] < 500), 'MinCategorize'] = 'less than 500'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 500) & (rides_minutes['TotalMinutes'] < 2000), 'MinCategorize'] = '500 to 1999'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 2000) & (rides_minutes['TotalMinutes'] < 4000), 'MinCategorize'] = '2000 to 3999'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 4000) & (rides_minutes['TotalMinutes'] < 7500), 'MinCategorize'] = '4000 to 7499'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 7500) & (rides_minutes['TotalMinutes'] < 10000), 'MinCategorize'] = '7500 to 9999'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 10000) & (rides_minutes['TotalMinutes'] < 15000), 'MinCategorize'] = '10000 to 14999'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 15000) & (rides_minutes['TotalMinutes'] < 25000), 'MinCategorize'] = '15000 to 24999'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 25000) & (rides_minutes['TotalMinutes'] < 50000), 'MinCategorize'] = '25000 to 49999'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 50000) & (rides_minutes['TotalMinutes'] < 100000), 'MinCategorize'] = '50000 to 99999'
rides_minutes.loc[(rides_minutes['TotalMinutes'] >= 100000), 'MinCategorize'] = '100000+'
```

```python
rides_minutes
```

```python
#saves the new dataframe to a new csv called minutes_categorize.csv
rides_minutes.to_csv('minutes_categorize.csv', sep=',')
```

```python
#loads the trip data csv into a dataframe

trip_data = pd.read_csv("C:/Users/Vi/Desktop/CMPE 188/Bikeshare-master/Data/trip_data.csv")
```

```python
trip_data
```

```python
#Prints duration of trip data to find the min and max values
trip_data['Duration'].describe()
```

```python
#makes an empty column called DurCategorize

trip_data['DurCategorize'] = ""
```

```python
#Sets the 10 different categorizes depending on range

trip_data.loc[(trip_data['Duration'] >= 0) & (trip_data['Duration'] < 120), 'DurCategorize'] = 'A'
trip_data.loc[(trip_data['Duration'] >= 120) & (trip_data['Duration'] < 360), 'DurCategorize'] = 'B'
trip_data.loc[(trip_data['Duration'] >= 360) & (trip_data['Duration'] < 525), 'DurCategorize'] = 'C'
trip_data.loc[(trip_data['Duration'] >= 525) & (trip_data['Duration'] < 750), 'DurCategorize'] = 'D'
trip_data.loc[(trip_data['Duration'] >= 750) & (trip_data['Duration'] < 1000), 'DurCategorize'] = 'E'
trip_data.loc[(trip_data['Duration'] >= 1000) & (trip_data['Duration'] < 1500), 'DurCategorize'] = 'F'
trip_data.loc[(trip_data['Duration'] >= 1500) & (trip_data['Duration'] < 2000), 'DurCategorize'] = 'G'
trip_data.loc[(trip_data['Duration'] >= 2000) & (trip_data['Duration'] < 4000), 'DurCategorize'] = 'H'
trip_data.loc[(trip_data['Duration'] >= 4000) & (trip_data['Duration'] < 8000), 'DurCategorize'] = 'I'
trip_data.loc[(trip_data['Duration'] >= 8000), 'DurCategorize'] = 'J'
```
